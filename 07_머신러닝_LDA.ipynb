# 서울페이 전체 리뷰 LDA 만들기

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer 
import pyLDAvis.lda_model
import pandas as pd
from konlpy.tag import Okt
import re

okt = Okt()

data = pd.read_csv("07_data/구글플레이스토어_서울페이_리뷰.csv", index_col=0)
# data

review_list = []
for i, review in enumerate(data["content"]):
    # Step1. 정규표현식 전처리하기
    review = re.sub("[^a-zA-Z0-9가-힣\\s]", "", review)
    # Step2. 형태소 분석기
    result = okt.morphs(review)
    # Step3. 담기
    # print(f"{i}번째 리뷰: {result}")
    # print(f"{i}번째 리뷰: {' '.join(result)}")
    # word_list.append(result)
    review_list.append(" ".join(result))


vectorizer = CountVectorizer(
    max_df = 0.1,               # 전체 단어 등장 비율이 p 이상인 것만 사용할 것.
    min_df = 2,                 # 단어가 적어도 n개 이상인 것만 사용할 것.
    max_features = 1000,        # 상위 항목 n개만 사용할 것.
    ngram_range= (1,2)          # 단어의 조합
)

feature_vec = vectorizer.fit_transform(review_list)
feature_vec.shape

feature_names = vectorizer.get_feature_names_out()
# print(feature_names)


lda = LatentDirichletAllocation(n_components=5)
lda.fit(feature_vec)

# print(lda.components_.shape)
# print(lda.components_)

for idx, topic in enumerate(lda.components_):
    # print(f"# Topic {idx+1}:")
    topic_word_idx = topic.argsort()[::-1]
    top_idxs = topic_word_idx[:20]
    topics = [feature_names[idx] for idx in top_idxs]
    # print(" ".join(topics))  


 
pyLDAvis.enable_notebook()
vis = pyLDAvis.lda_model.prepare(lda, feature_vec, vectorizer)
pyLDAvis.display(vis)
```
